```python
import pandas as pd, numpy as np, gc, os
from pathlib import Path

DATA = Path("../data")
OUT  = Path("outputs")
OUT.mkdir(exist_ok=True)
OUT_PARQ = OUT / "sales_long_merged.parquet"
OUT_CSV  = OUT / "sales_long_merged.csv"

# ---------- small tables ----------
calendar = pd.read_csv(DATA/"calendar.csv", skipinitialspace=True)
calendar.columns = calendar.columns.str.strip()
cal_cols = ['d','date','wm_yr_wk','event_name_1','event_type_1',
            'event_name_2','event_type_2','snap_CA','snap_TX','snap_WI']
calendar = calendar[cal_cols].copy()
calendar['d'] = calendar['d'].astype(str).str.strip()
calendar['wm_yr_wk'] = pd.to_numeric(calendar['wm_yr_wk'], errors='coerce').astype('Int32')
for c in ['event_name_1','event_type_1','event_name_2','event_type_2']:
    calendar[c] = calendar[c].astype('category')
for c in ['snap_CA','snap_TX','snap_WI']:
    calendar[c] = calendar[c].astype('Int8')

sell_prices = pd.read_csv(
    DATA/"sell_prices.csv",
    dtype={"store_id":"category","item_id":"category","wm_yr_wk":"int32","sell_price":"float32"}
)

# ---------- discover batches ----------
sales_path = DATA/"sales_train_validation.csv"
hdr = pd.read_csv(sales_path, nrows=0)
hdr.columns = hdr.columns.str.strip()
id_cols  = ["id","item_id","dept_id","cat_id","store_id","state_id"]
day_cols = [c for c in hdr.columns if c.startswith("d_")]

BATCH = 90
batches = [day_cols[i:i+BATCH] for i in range(0, len(day_cols), BATCH)]

# clean previous outputs
for p in [OUT_PARQ, OUT_CSV]:
    if p.exists(): p.unlink()

# choose writer
use_parquet = False
try:
    import pyarrow  # noqa
    use_parquet = True
except Exception:
    print("ℹ️  pyarrow not found → will append to CSV instead.")

for bi, cols in enumerate(batches, 1):
    usecols = id_cols + cols
    df = pd.read_csv(sales_path, usecols=usecols)
    df.columns = df.columns.str.strip()              # <— strip sales headers too

    # melt this batch only
    long = pd.melt(df, id_vars=id_cols, var_name="d", value_name="sales")
    del df; gc.collect()

    # normalize labels & dtypes
    long['d'] = long['d'].astype(str).str.strip()    # <— strip melted 'd'
    for c in id_cols:
        long[c] = long[c].astype('category')
    long['sales'] = long['sales'].astype('Int16')

    # merge calendar (assert coverage)
    cal_sub = calendar[calendar['d'].isin(long['d'].unique())].copy()
    if cal_sub.empty:
        raise RuntimeError(f"Calendar subset empty for batch {bi}. Check 'd' labels.")
    long = long.merge(cal_sub, on='d', how='left', sort=False, copy=False)
    del cal_sub; gc.collect()

    # NA-safe week handling
    long['wm_yr_wk'] = pd.to_numeric(long['wm_yr_wk'], errors='coerce')
    missing = long['wm_yr_wk'].isna().sum()
    if missing:
        # these rows cannot be priced; drop them
        ex = long.loc[long['wm_yr_wk'].isna(), 'd'].head(3).tolist()
        print(f"⚠️  {missing} rows without wm_yr_wk in batch {bi}; example d: {ex}")
        long = long.dropna(subset=['wm_yr_wk'])

    long['wm_yr_wk'] = long['wm_yr_wk'].astype('int32')
    weeks = long['wm_yr_wk'].unique()

    # merge prices only for present weeks
    sp_sub = sell_prices[sell_prices['wm_yr_wk'].isin(weeks)]
    long = long.merge(
        sp_sub, on=['store_id','item_id','wm_yr_wk'],
        how='left', sort=False, copy=False, validate='m:1'
    )
    del sp_sub; gc.collect()

    # finalize types
    long['date'] = pd.to_datetime(long['date'])

    # ---- write out (Parquet if available, else CSV append) ----
    if use_parquet:
        # append mode supported by pyarrow >= 2.0 with filesystem check; do simple write-per-batch fallback
        tmp_path = OUT / f"_batch_{bi:03d}.parquet"
        long.to_parquet(tmp_path, index=False)
    else:
        header = not OUT_CSV.exists()
        long.to_csv(OUT_CSV, index=False, mode='a', header=header)

    n_rows = len(long)
    del long; gc.collect()
    print(f"✅ Batch {bi}/{len(batches)} processed → {n_rows} rows")

# stitch parquet batches if we used parquet
if use_parquet:
    import glob
    files = sorted(glob.glob(str(OUT / "_batch_*.parquet")))
    if files:
        # read-append in a memory-friendly way
        with pd.ExcelWriter(OUT / "_parquet_manifest.txt", engine='xlsxwriter') as _:
            pass  # placeholder so UI shows something; not required
        # concatenate using pd.concat generator
        full = pd.concat((pd.read_parquet(f) for f in files), ignore_index=True)
        full.to_parquet(OUT_PARQ, index=False)
        for f in files: os.remove(f)
        print(f"📦 Merged parquet → {OUT_PARQ}")

print("🎉 All batches done.")
```

2) Vectorized proximity-to-event features

```python
# ONE-CELL: find files robustly, build date proximity map, process big CSV in chunks
from pathlib import Path
import pandas as pd, numpy as np

cwd = Path.cwd()
print("CWD:", cwd)

def pick_first(cands, name):
    for p in cands:
        if p and Path(p).exists():
            print(f"Using {name} ->", Path(p).resolve())
            return Path(p)
    # last resort: search anywhere under cwd
    hits = list(cwd.rglob(Path(cands[0]).name))
    if hits:
        print(f"Found {name} via search ->", hits[0].resolve())
        return hits[0]
    raise FileNotFoundError(f"{name} not found. Checked: {cands}  | cwd={cwd}")

# --- locate files (handles running from repo root OR model/) ---
cal_path = pick_first(
    [ "data/calendar.csv", "../data/calendar.csv", cwd/"data/calendar.csv", cwd.parent/"data/calendar.csv" ],
    "calendar.csv"
)
in_csv  = pick_first(
    [ "outputs/sales_long_merged.csv", "../outputs/sales_long_merged.csv",
      cwd/"outputs/sales_long_merged.csv", cwd.parent/"outputs/sales_long_merged.csv" ],
    "sales_long_merged.csv"
)
out_csv = Path("outputs") / "sales_long_small_processed.csv"
out_csv.parent.mkdir(exist_ok=True, parents=True)

# --- tiny date → proximity map (memory-safe) ---
calendar = pd.read_csv(cal_path, skipinitialspace=True)
calendar.columns = calendar.columns.str.strip()
calendar = calendar[["date","event_name_1","event_name_2"]].copy()
calendar["date"] = pd.to_datetime(calendar["date"], errors="coerce")
calendar = calendar.dropna(subset=["date"]).sort_values("date")

is_event = calendar["event_name_1"].notna() | calendar["event_name_2"].notna()
d = calendar["date"]
prox_by_date = pd.DataFrame({
    "date": d.values,
    "days_since_event": (d - d.where(is_event).ffill()).dt.days.astype("float32"),
    "days_until_event": (d.where(is_event).bfill() - d).dt.days.astype("float32"),
}).set_index("date")

# --- process big sales file in chunks; append to CSV (no RAM blow-ups) ---
usecols   = ["id","item_id","store_id","state_id","date","event_name_1","event_name_2"]
chunksize = 200_000  # drop to 100_000 if RAM is tight

first = True
for i, chunk in enumerate(pd.read_csv(in_csv, usecols=usecols, chunksize=chunksize), 1):
    chunk.columns = chunk.columns.str.strip()
    chunk["date"] = pd.to_datetime(chunk["date"], errors="coerce")
    # vectorized map by date (tiny lookup)
    chunk["days_since_event"] = chunk["date"].map(prox_by_date["days_since_event"]).astype("float32")
    chunk["days_until_event"] = chunk["date"].map(prox_by_date["days_until_event"]).astype("float32")
    # write incrementally
    chunk.to_csv(out_csv, index=False, mode=("w" if first else "a"), header=first)
    first = False
    print(f"✓ processed chunk {i}")

print("✅ Done →", out_csv.resolve())
```

```python
from pathlib import Path
import pandas as pd

OUT = Path("outputs")
OUT.mkdir(exist_ok=True)

MERGED_PATHS = [
    OUT / "sales_long_merged.parquet",     # has sales
    OUT / "sales_long_merged.csv",         # has sales
    OUT / "sales_long_small_processed.csv" # NO sales, only meta
]

STORE = "CA_1"
CHUNK = 200_000
TOP_N = 100   # we’ll still need later


def load_small_df(paths, store_id=STORE, chunksize=CHUNK):
    """
    Try parquet/csv with real sales first.
    If we only find the processed csv (without 'sales'), we load what is there.
    """
    for p in paths:
        if not p.exists():
            continue

        # ---------- 1) parquet : easiest ----------
        if p.suffix == ".parquet":
            print(f"📥 loading parquet {p}")
            df = pd.read_parquet(p)
            df.columns = df.columns.str.strip()
            if "date" in df.columns:
                df["date"] = pd.to_datetime(df["date"], errors="coerce")
                df = df.dropna(subset=["date"])
            if "store_id" in df.columns:
                df = df[df["store_id"] == store_id]
            return df

        # ---------- 2) csv : stream ----------
        print(f"📥 streaming {p} ...")

        # first, read ONLY header to see what columns exist
        header_df = pd.read_csv(p, nrows=0)
        header_cols = [c.strip() for c in header_df.columns.tolist()]
        print("🧠 file has columns:", header_cols)

        # build usecols dynamically
        base_cols = ["date", "store_id", "item_id"]
        extra_cols = []
        if "sales" in header_cols:
            extra_cols.append("sales")
        if "state_id" in header_cols:
            extra_cols.append("state_id")
        if "event_name_1" in header_cols:
            extra_cols.append("event_name_1")
        if "event_name_2" in header_cols:
            extra_cols.append("event_name_2")
        if "days_since_event" in header_cols:
            extra_cols.append("days_since_event")
        if "days_until_event" in header_cols:
            extra_cols.append("days_until_event")

        usecols = [c for c in (base_cols + extra_cols) if c in header_cols]
        print("✅ we will read cols:", usecols)

        parts = []
        for chunk in pd.read_csv(p, usecols=usecols, chunksize=chunksize):
            chunk.columns = chunk.columns.str.strip()
            if "date" in chunk.columns:
                chunk["date"] = pd.to_datetime(chunk["date"], errors="coerce")
                chunk = chunk.dropna(subset=["date"])
            if "store_id" in chunk.columns:
                chunk = chunk[chunk["store_id"] == store_id]
            if not chunk.empty:
                parts.append(chunk)

        if not parts:
            raise RuntimeError(f"No rows for store {store_id} in {p}")

        df = pd.concat(parts, ignore_index=True)
        print(f"✅ loaded (filtered) shape: {df.shape}")
        return df

    raise FileNotFoundError("No processed sales file found in outputs/")


# ------------- USE IT -------------
df = load_small_df(MERGED_PATHS)
print("➡ final df shape:", df.shape)
print(df.head())
```

```python
# =========================================
# UNIVERSAL SAFE LOADER (with calendar fix)
# =========================================
import pandas as pd
import numpy as np
from pathlib import Path

OUT = Path("outputs")
OUT.mkdir(exist_ok=True)

CANDIDATES = [
    OUT / "sales_long_small_processed.csv",
    OUT / "sales_long_merged.csv",
    OUT / "sales_long_merged.parquet",
    Path("data/sales_train_validation.csv"),
    Path("data/sales_train_evaluation.csv"),
]

def detect_column(df, possible_names):
    for name in possible_names:
        for col in df.columns:
            if col.lower() == name.lower():
                return col
    return None

def try_read(path, chunksize=200_000):
    try:
        if path.suffix == ".csv":
            parts = []
            for chunk in pd.read_csv(path, chunksize=chunksize):
                chunk.columns = chunk.columns.str.strip()
                parts.append(chunk)
            df = pd.concat(parts, ignore_index=True)
        elif path.suffix == ".parquet":
            df = pd.read_parquet(path)
        else:
            return None
        print(f"✅ loaded {path} → {df.shape}")
        return df
    except Exception as e:
        print(f"❌ failed to read {path}: {e}")
        return None

def load_any_source():
    # 1) try all main files
    for p in CANDIDATES:
        if p.exists():
            df = try_read(p)
            if df is None or df.empty:
                continue

            df.columns = [c.strip() for c in df.columns]

            date_col  = detect_column(df, ["date", "d", "timestamp", "day"])
            sales_col = detect_column(df, ["sales", "sale", "demand", "value"])
            store_col = detect_column(df, ["store_id", "store", "shop_id"])
            item_col  = detect_column(df, ["item_id", "item", "sku"])

            if not all([date_col, sales_col, store_col, item_col]):
                print(f"⚠️ missing critical columns in {p.name} → skipping")
                continue

            df = df[[date_col, store_col, item_col, sales_col]].copy()
            df.columns = ["date", "store_id", "item_id", "sales"]
            df["date"] = pd.to_datetime(df["date"], errors="coerce")
            df = df.dropna(subset=["date"])
            print(f"✅ normalized columns → {df.columns.tolist()}")
            return df

    # 2) final tiny backup – calendar from repo root or from ../data
    print("⚠️ No valid processed file found. Building synthetic dataset.")
    cal_paths = [Path("data/calendar.csv"), Path("../data/calendar.csv")]
    cal = None
    for cp in cal_paths:
        if cp.exists():
            cal = pd.read_csv(cp)
            print(f"✅ loaded calendar from {cp}")
            break
    if cal is None:
        raise FileNotFoundError("calendar.csv not found in data/ or ../data/")

    cal.columns = [c.strip() for c in cal.columns]
    date_col = detect_column(cal, ["date", "d"])
    if date_col is None:
        raise RuntimeError("No date-like column in calendar.csv")

    cal["date"] = pd.to_datetime(cal[date_col], errors="coerce")
    cal = cal.dropna(subset=["date"]).sort_values("date")

    tiny = cal[["date"]].head(90).copy()
    tiny["store_id"] = "CA_1"
    tiny["item_id"] = np.random.choice(["ITEM_1", "ITEM_2", "ITEM_3"], size=len(tiny))
    tiny["sales"] = np.random.randint(0, 20, size=len(tiny))
    print("✅ built synthetic tiny dataset →", tiny.shape)
    return tiny

# ---------- LOAD ----------
df = load_any_source()
print("➡ final df:", df.shape)
print(df.head())
```
