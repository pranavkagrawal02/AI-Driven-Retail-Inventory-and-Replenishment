```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
sample_submission_dataframe = pd.read_csv("../data/sample_submission.csv")
ssdf=sample_submission_dataframe
p = Path.cwd()
while not (p / "data").exists() and p != p.parent:
    p = p.parent
DATA_DIR = p / "data"          # .../AIretail/data
print("\nLoaded from:", DATA_DIR)
print("\nColumns:", ssdf.columns.tolist())
print("\nFirst row:", ssdf.iloc[0].tolist())
labels = ssdf.columns.tolist()
sizes = []
for i in range(len(ssdf.columns.tolist())):
    sizes.append(1)
print("\nColoums in sample_submission_dataset.csv Dataset")
for i in range(len(ssdf.columns.tolist())):
    print(i+1,": ",ssdf.columns.tolist()[i])
plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90)
plt.axis('equal')  # Ensures circular pie
plt.title('Pie Chart: Coloums in sample_submission_dataset.csv Dataset')
plt.show()
```
