```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
# Read CSV correctly (fix spelling)
calender_dataframe = pd.read_csv("../data/calendar.csv")

caldf=calender_dataframe

p = Path.cwd()
while not (p / "data").exists() and p != p.parent:
    p = p.parent
DATA_DIR = p / "data"          # .../AIretail/data

print("\nLoaded from:", DATA_DIR)
print("\nColumns:", caldf.columns.tolist())
print("\nFirst row:", caldf.iloc[0].tolist())

labels = caldf.columns.tolist()

sizes = []

for i in range(len(caldf.columns.tolist())):
    sizes.append(1)

'''
print(sizes)

print("sizes length: ", len(sizes))
'''
for i in range(len(caldf.columns.tolist())):
    print(i+1,": ",caldf.columns.tolist()[i])

plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90)
plt.axis('equal')  # Ensures circular pie
plt.title('Coloums in Calendar.csv Dataset')
plt.show()
```
