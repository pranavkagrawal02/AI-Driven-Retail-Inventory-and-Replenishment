```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

sales_train_evaluation_dataframe = pd.read_csv("../data/sales_train_evaluation.csv")



stedf=sales_train_evaluation_dataframe

p = Path.cwd()
while not (p / "data").exists() and p != p.parent:
    p = p.parent
DATA_DIR = p / "data"          # .../AIretail/data

print("\nLoaded from:", DATA_DIR)
print("\nColumns:", stedf.columns.tolist())
print("\nFirst row:", stedf.iloc[0].tolist())

labels = stedf.columns.tolist()

sizes = []

for i in range(len(stedf.columns.tolist())):
    sizes.append(1)

'''
print(sizes)

print("sizes length: ", len(sizes))
'''
print("\nColoums in sales_train_evaluation.csv Dataset")
for i in range(len(stedf.columns.tolist())):
    print(i+1,": ",stedf.columns.tolist()[i])



plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90)
plt.axis('equal')  # Ensures circular pie
plt.title('Pie Chart: Coloums in sales_train_evaluation.csv Dataset')
plt.show()
```
