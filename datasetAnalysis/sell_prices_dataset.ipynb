```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
sell_prices_dataframe = pd.read_csv("../data/sell_prices.csv")
spdf=sell_prices_dataframe
p = Path.cwd()
while not (p / "data").exists() and p != p.parent:
    p = p.parent
DATA_DIR = p / "data"          # .../AIretail/data
print("\nLoaded from:", DATA_DIR)
print("\nColumns:", spdf.columns.tolist())
print("\nFirst row:", spdf.iloc[0].tolist())
labels = spdf.columns.tolist()
sizes = []
for i in range(len(spdf.columns.tolist())):
    sizes.append(1)
print("\nColoums in sell_price_dataset.csv Dataset")
for i in range(len(spdf.columns.tolist())):
    print(i+1,": ",spdf.columns.tolist()[i])
plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90)
plt.axis('equal')  # Ensures circular pie
plt.title('Pie Chart: Coloums in ssell_price_dataset.csv Dataset:')
plt.show()
```
